```python
# 문제 3. 이슬점 온도와 상관관계(correlation)가 있는 기상정보는 무엇일지 알아보고자 합니다. 
# 이를 위해 상관계수(pearson correlation coefficient)를 계산해보고, 
# 산점도(Scatter plot)를 그려봅시다.
import numpy as np

def corr(u1, u2):
    # mean()->비어있지 않은 값들의 평균을 구해줌
    # 평균값을 빼서 축을 옮겨주는 역할
    u1_c = u1 - u1.mean() 
    u2_c = u2 - u2.mean()
    
    
    denom = np.sqrt(np.sum(u1_c ** 2) * np.sum(u2_c ** 2))
    
    if denom != 0:
        return np.sum(u1_c * u2_c)/denom
    else:
        return 0
    
```

```python
a = corr(data['습도'], data['이슬점 온도'])
b = corr(data['풍속'], data['이슬점 온도'])
c = corr(data['현지기압'], data['이슬점 온도'])
d = corr(data['기온'], data['이슬점 온도'])

print(a, b, c, d)
```

```python
# 이슬점 온도 구하는 식을 Regression 기법을 통해 구해봅시다.
# 이슬점 온도 구하는 식은 미지수 a, b, c, d, e에 대해서 아래와 같다고 가정합니다.

# 이슬점 온도 = a * 습도 + b * 풍속 + c * 현지기압 + d * 기온 + e

# 이 때 a, b, c, d, e 값을 구하고, 각 값을 콜론(:)으로 구분하여 제출하세요.
# (예: a=1, b=2, c=3, d=4, e=5라면, "1:2:3:4:5"를 제출)

from sklearn.linear_model import LinearRegression

model = LinearRegression()

X = data[['습도','풍속', '현지기압', '기온' ]]
Y = data['이슬점 온도']

# print(X, Y)
model.fit(X, Y)

print(model.coef_)
print(model.intercept_)

# 정답 : 0.2676329 : -0.03078923 :  0.00952646 : 0.92936961 : -32.91662937401235
```

```python
# 예측결과
predicted = model.predict(X)

# print(predicted)
# print(predicted.shape)
# print(Y.shape)

((predicted - Y) **2).mean()
```

```python

def cosine(v1, v2):
    return (v1*v2).sum() / (((v1*v1).sum() * (v2 * v2).sum()) ** 0.5)

cosine(data['train_vectors'][0], data['test1_vectors'][1])
```

```python
ids = []

for i in range(10) :
  arr = []
  for j in range(data['train_vectors'].shape[0]):
    arr.append((cosine(data['test1_vectors'][i], data['train_vectors'][j]), j))


  arr.sort(reverse=True)
  ids.append([i for c, i in arr[:3]])

print(ids)
```

```python
# 아래는 data['train_vectors'] 와 data['train_labels']를 학습하도록, Softmax Regression을 pytorch로 구현한 것입니다. 
# 빈 칸에 들어갈 문구를 순서대로 콜론(:)으로 구분하여 제출하세요. (답 사이에 공백을 넣지 마세요).

import torch
import torch.nn as nn
import torch.nn.functional as F

x_train = torch.from_numpy(data["train_vectors"])
y_train = torch.from_numpy(data["train_labels"])

x = torch.unsqueeze(x_train, 2)
y = torch.unsqueeze(y_train, 1)

model = nn.Linear(256, 3)

optim = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(10001):
    z = model(x_train)
    cost = F.cross_entropy(z, y_train)

    optim.zero_grad()
    cost.backward()
    optim.step()

    with torch.no_grad():
        if epoch % 1000 == 0:
            print(f"epoch: {epoch}, cost: {cost.item()}")
```
