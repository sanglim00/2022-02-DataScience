```python
from sklearn.cluster import KMeans
import pickle as pkl

data = pkl.load(open("final_dataset1.pkl", "rb"))

X = []
Y = []

for k in range(1, 16):
  kmeans = KMeans(n_clusters = k)
  kmeans.fit(data)

  print(kmeans.inertia_)
  X.append(k)
  Y.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.plot(X, Y, "-o")
```
